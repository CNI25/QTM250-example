#INTRODUCTION
Speech to text recognition software allows for the transcription of spoken languages into text. This is done through computational linguistics: "linguistic algorithms [are used] to sort auditory signals and convert them to words, which are then displayed as Unicode characters." Speech-to-text has many uses in real life, such as interview or lecture transcription and tools such as Siri and Alexa. This project explores the accuracy of speech-to-text software depending on the distance of a selected video from the machine.

#HYPOTHESIS
Our hypothesis is that the closer the audio is to the machine, the more accurate the transcription will be. We think that abbreviations, as well as music or speech with background noise, will be harder to transcribe.

#METHODS
We chose four distances in feet (0, 1, 3, 6) and three different scripts with various linguistic components such as possessives (i.e."'s"), abbreviations (i.e. "you're"), and numbers to generate a total of 12 samples.



> These are the three audio bits we decided to record, with one group member reading all of them for consistency: 
- "Refusing to go to the gym is one of the best forms of resistance training."
- "New York's housing affordability crisis: Person making minimum wage would need to work 99 hours to afford a one-bedroom apartment."
- "If you’re the kind of person that has no good luck, when you see the light at the end of the tunnel… run, because the train is coming."



Before we began data collection, the first thing we did was to make an audio recording of each script, so that confounding variables such as volume and speed of the speech can be avoided. When collecting data, we set a recording device in one place and played the audio at the four different distanecs. Then, we entered the recordings one-by-one into Shell to generate transcripts using the Google Cloud Speech API, as shown by the screenshot below.

Then, we recorded our data into a Google Spreadsheet with four columns - script ID, distance, transcription, confidence. To further analyze our data, we counted the number of correctly transcribed words and calculated the percentage of each. Below is our data table.

#RESULTS

Overall, the Speech-to-Text API generates pretty accurate results. The average confidence level across all samples at all distances was 95.34. However, when looking at confidence level by distance, some samples were more accurate than others. Similar to our hypothesis, samples collected closer to the device generated transcripts with higher confidence levels. As shown by the first line graph below, there is a negative correlation between distance and confidence. However, it is interesting to note that, on average, when the audio is placed 1 feet away from the computer, the transcript tends to be more accurate than when the audio is played right above the computer with a distance of zero.
Confidence by distance.png

When looking at the number and proportion of correctly transcribed words, there doesn't seem to be as much of difference by distance. The drop in number for 3-feet distance was most likely due to an outlier sample which failed to be transcribed.
